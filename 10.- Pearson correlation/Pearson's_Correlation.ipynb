```python
import pandas as pd
import numpy as np
```

```python
# Leemos dataset
data = pd.read_csv("Sales_200_days.csv")
```

```python
# Explorar el dataset
data.head()
```

#**Pearson's Correlation Coefficient**

**Combination: Newspaper-Sales**

```python
# Calculate Pearson's correlation coefficient
correlation_matrix = np.corrcoef(data["Newspaper"], data["Sales"])

correlation_matrix
```

```python
# The correlation coefficient between x and y is located at position (0, 1) or (1, 0) in the matrix
correlation_coefficient = correlation_matrix[0, 1]

#Create new column to assign the corelation value
data["Newspaper-Sales"] = correlation_coefficient

print("Pearson's correlation coefficient for the inputs Newspaper-Sales:", correlation_coefficient)
```

**Combination: Radio-Sales**

```python
# Calculate Pearson's correlation coefficient
correlation_matrix = np.corrcoef(data["Radio"], data["Sales"])

correlation_matrix
```

```python
# The correlation coefficient between x and y is located at position (0, 1) or (1, 0) in the matrix
correlation_coefficient = correlation_matrix[0, 1]

#Create new column to assign the corelation value
data["Radio-Sales"] = correlation_coefficient



print("Pearson's correlation coefficient for the inputs Radio-Sales:", correlation_coefficient)
```

**Combination: TV-Sales**

```python
# Calculate Pearson's correlation coefficient
correlation_matrix = np.corrcoef(data["TV"], data["Sales"])

correlation_matrix
```

```python
# The correlation coefficient between x and y is located at position (0, 1) or (1, 0) in the matrix
correlation_coefficient = correlation_matrix[0, 1]

#Create new column to assign the corelation value
data["TV-Sales"] = correlation_coefficient

print("Pearson's correlation coefficient for the inputs TV-Sales:", correlation_coefficient)
```

#**Conclusion**

Pearson correlation coeficient, provides important information about the relationship between two variables, it reflets if the relationship is positive or negative and also gives information about the strength of the connection between the variables. In this example the variables TV-Sales are the ones that provides higher correlation with a positive 0.78 value, the combination of the variables Radio-Sales provides a medium correlation with 0.57 and the variables newspaper-Sales, give a low positive correlation with a value of 0.22. In conclusion the variable TV is the one with major impact on the dependent variable, in this scenario the decision makers could take the decision to invest in a marketing strategy using the TV as a main broadcast media.
