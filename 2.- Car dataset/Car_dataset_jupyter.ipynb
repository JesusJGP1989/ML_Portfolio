```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt  

#Cargar el dataset. utilizaremos como depósito del dataset una estructura bidimencional
df = pd.read_csv('imports-85.csv')

# Imprimir los primeros 5 registros de datos
df.head()
```

```python
# Imprimir los últimos 5 registros de datos
df.tail()
```

```python
# Agregar aencabezados a las columnas 
headers= ["symboling","normalized-losses","make","fuel-type","aspiration", "num-of-doors","body- style", "drive-wheels","engine-location","wheel-base",
"length","width","height","curb-weight","engine-type", "num-of-cylinders", "engine-size","fuel-system","bore","stroke","compression-ratio","horsepower", "peak- rpm","city-mpg","highway-mpg","price"]
df.columns = headers

# Imprimir los primeros 5 registros de datos
df.head()
```

```python
#Remplazar los datos que contienen el valor "?"
df.replace("?", np.nan)
```

```python
#Convertir los titulos de las columnas "city-mpg", "highway-mpg" a "city-kpl", "highway-kpl".

conversion_factor = 0.425144

df['city-kpl'] =df['city-mpg']*conversion_factor
df['highway-kpl']=df['highway-mpg']*conversion_factor

# Imprimir los primeros 5 registros de datos
df.head()
```
