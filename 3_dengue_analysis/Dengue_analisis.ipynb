```python
# Importamos la librería Pandas
import pandas as pd
# # LIbrería para análisis numérico
import numpy as np
# Libreria para crear gráficas
import matplotlib.pyplot as plt

import seaborn as sns
```

```python
# Cargar el dataset
df = pd.read_csv("dengue_abierto.csv")
```

```python
#Remplazar los datos que contienen el valor "?"
df.replace("?", np.nan)
```

```python
#Check and handle missing data if necessary
missing_data = df.isnull().sum()
print("Missing data:\n", missing_data)
```

```python
#Fill empty fields
# Replace all NaN values with a specified value, for example, 0
df_filled = df.fillna(0)

#Check and handle missing data if necessary
missing_data = df_filled.isnull().sum()
print("Missing data:\n", missing_data)
```

```python
# Print statistical information about the data
print(df_filled.describe())
```

```python
#Datatypes
print(df_filled.dtypes)

# Identify categorical variables
categorical_variables = df_filled.select_dtypes(include=['object', 'category']).columns.tolist()
```

```python
# obtain the three measures of central tendency from the numerical cells of the dataset.
measure = 'EDAD_ANOS'

# Calculate metrics
average = df_filled[measure].mean()  # Arithmetic mean
fashion = df_filled[measure].mode()  # Mode
std_deviation = df_filled[measure].std()  # Standard deviation

# Print the obtained metrics
print("Arithmetic Mean:", average)
print("Mode (Fashion):", fashion.iloc[0])
print("Standard Deviation:", std_deviation)
```

```python
#Count the number of cases per category for each categorical variable.
# Categories: men, women, children, adolescents, adults and older adults
OlderAdults = df_filled[df_filled['EDAD_ANOS']>60]
Adults = df_filled[ (df_filled['EDAD_ANOS']>18) & (df['EDAD_ANOS']<=60) ]
Adolescents = df_filled[ (df_filled['EDAD_ANOS']>=12) & (df['EDAD_ANOS']<=18) ]
Children = df_filled[ (df_filled['EDAD_ANOS']>=0) & (df['EDAD_ANOS']<12) ]
women = df_filled[df_filled['SEXO']== 1]
man   = df_filled[df_filled['SEXO']== 2]

indigena     = df_filled[df_filled['INDIGENA']== 1]
no_indigena  = df_filled[df_filled['INDIGENA']== 2]

print('Older adults: {}'.format(len(OlderAdults)))
print('Adults: {}'.format(len(Adults)))
print('Adolescents: {}'.format(len(Adolescents)))
print('Children: {}\n'.format(len(Children)))

print('women: {}'.format(len(women)))
print('man: {}\n'.format(len(man)))

print('Indigenous: {}'.format(len(indigena)))
print('No indigenous: {}\n'.format(len(no_indigena)))
```

```python
#8#Obtain the number of cases of classic dengue and dengue hemorrhagic fever

hemorragico  = df_filled[df_filled['HEMORRAGICOS']== 1]
no_hemorragico  = df_filled[df_filled['HEMORRAGICOS']== 2]


print('Hemorragic dengue: {}'.format(len(hemorragico)))
print('No hemorragic dengue: {}'.format(len(no_hemorragico)))
```

```python
#9# Obtain the number of patients who died from each of the two types of dengue and by age category.
OlderAd_hemorragic_death =  OlderAdults[ (OlderAdults['HEMORRAGICOS']==1) & (OlderAdults['DEFUNCION']==1) ]
Adults_hemorragic_death =  Adults[ (Adults['HEMORRAGICOS']==1) & (Adults['DEFUNCION']==1) ]
Adole_hemorragic_death =  Adolescents[ (Adolescents['HEMORRAGICOS']==1) & (Adolescents['DEFUNCION']==1) ]
Child_hemorragic_death =  Children[ (Children['HEMORRAGICOS']==1) & (Children['DEFUNCION']==1) ]

OlderAd_no_hemorragic_death =  OlderAdults[ (OlderAdults['HEMORRAGICOS']==2) & (OlderAdults['DEFUNCION']==1) ]
Adults_no_hemorragic_death =  Adults[ (Adults['HEMORRAGICOS']==2) & (Adults['DEFUNCION']==1) ]
Adole_no_hemorragic_death =  Adolescents[ (Adolescents['HEMORRAGICOS']==2) & (Adolescents['DEFUNCION']==1) ]
Child_no_hemorragic_death =  Children[ (Children['HEMORRAGICOS']==2) & (Children['DEFUNCION']==1) ]

print('Older adults hemorragic deaths :{} '.format(len(OlderAd_hemorragic_death) ) )
print('Adults hemorragic deaths :{} '.format(len(Adults_hemorragic_death) ) )
print('Adolescent hemorragic deaths :{} '.format(len(Adole_hemorragic_death) ) )
print('Children hemorragic deaths :{}\n'.format(len(Child_hemorragic_death) ) )

print('Older adults no hemorragic deaths :{} '.format(len(OlderAd_no_hemorragic_death) ) )
print('Adults no hemorragic deaths :{} '.format(len(Adults_no_hemorragic_death) ) )
print('Adolescent no hemorragic deaths :{} '.format(len(Adole_no_hemorragic_death) ) )
print('Children no hemorragic deaths :{} '.format(len(Child_no_hemorragic_death) ) )
```

```python
#10#. Visualize the distribution of dengue cases by category using bar graphs.
Labels = ['Older adults', 'Adults', 'Adolescents', 'Children']

Datos = [ len(OlderAdults), len(Adults), len(Adolescents), len(Children) ]

plt.bar(Labels, Datos, color='blue')

plt.title('Distribution of dengue cases by age category')
plt.xlabel('Category')
plt.ylabel('Values')

#Mostrar gráfica
plt.grid(True, linestyle='--', alpha=0.7)  # Agregar una cuadrícula
plt.tight_layout()  # Ajustar diseño
plt.show()


```

```python
#Visualize the number of dengue cases over time using a line graph or bar graph grouped by month

# Gráfica de distribución de marca de automóviles
# Obtener el número de autos  de cada marca
DateSymptom = df_filled["FECHA_SIGN_SINTOMAS"].value_counts()

# Graficar la distribución del numero de casos en un gráfico de barras
ax = DateSymptom.plot(kind="bar", figsize=(25, 6), color='skyblue', edgecolor='black', fontsize=12)
DateSymptom.plot(kind="bar", figsize=(25, 6), color='skyblue', edgecolor='black', fontsize=12)

# Agregar título y etiquetas a los ejes
plt.title("Numer of registered dengue symptoms per day")
plt.xlabel("Date")
plt.ylabel("Cases")

# Agregar la cantidad de casos por día en cada barra
for i, v in enumerate(DateSymptom):
   ax.text(i, v + 0.2, str(v), ha='center', va='bottom')

# Mostrar la gráfica
# Rotar las etiquetas del eje x para mayor claridad
plt.xticks(rotation=45)

# Agregar una cuadrícula horizontal
plt.grid(axis='y', linestyle='--', alpha=0.7)
# Ajustar diseño
plt.tight_layout()
plt.show()
```

```python
#Explore the correlation between at least two variables from the dataset columns in the dataset
# Calculate correlation coefficient
correlationEntidad = df_filled['ENTIDAD_UM_NOTIF'].corr(df_filled['ENTIDAD_ASIG'])
print('Correlation between notified estate vs assigned estate to health users:{}\n '.format(correlationEntidad))

correlationPadecimiento = df_filled['DIABETES'].corr(df_filled['HIPERTENSION'])
print('Correlation between people suffering from diabetes and hypertension:{}\n'.format(correlationPadecimiento))

correlationHemorragic = (df_filled['HEMORRAGICOS']==1).corr(df_filled['DEFUNCION']==1)
print('Correlation between dengue hemorrhagic fever and deaths:{}\n '.format(correlationHemorragic))

Correlation_list = [correlationEntidad, correlationPadecimiento, correlationHemorragic]


# Create heatmap 1
plt.figure(figsize=(6, 4))
sns.heatmap([Correlation_list], annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title('Heatmap: Correlation ')
plt.xlabel('VARIABLES')
plt.ylabel('COEFICIENT')
plt.show()

```
