```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt  


#Cargar los datasets
df1 = pd.read_csv('dataset1.csv')
df2 = pd.read_csv('dataset2.csv')
df3 = pd.read_csv('dataset3.csv')
df4 = pd.read_csv('dataset4.csv')
df5 = pd.read_csv('dataset5.csv')


#Merge dataframes 
merged_df= pd.concat([df1, df2, df3, df4, df5], ignore_index=True)

#Agrega nombres a las columnas 
headers = ['Temperatura', 'humedad', 'Humedad terreno', 'Luminosidad','Ph', 'Velocidad viento']
merged_df.columns = headers

# Imprimir los primeros 5 registros de datos
merged_df.head()
```

```python
#agrega informaci√≥n estadistica
df_stats = merged_df.describe()

print(df_stats)
```

```python
#Obtener los valores maximos y minimos
min_value = df_stats.loc['min']
max_value = df_stats.loc['max']

# Calcula el rango
data_range = max_value - min_value

print(data_range)
```

```python
#standar deviation 
std_deviation = merged_df.std()
print(std_deviation)
```

```python
#variance
print(merged_df.var())
```

```python
#Interquartile Range (IQR)
q1 = merged_df.quantile(0.25)
q3 = merged_df.quantile(0.75)

# Calculate the Interquartile Range (IQR)
iqr = q3 - q1

print(iqr)
```

```python
#Coefficient of variation (CV)
mean_value = merged_df.mean()
std_deviation = merged_df.std()

cv = (std_deviation/mean_value)*100

print(cv)
```
