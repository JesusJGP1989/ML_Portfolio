## Merge, Join, and Concatenation Operations in Datasets

### Introduction
When working with datasets, it's often necessary to combine or merge data from different sources. This process can be achieved through various operations like merge, join, and concatenation. Each operation serves a specific purpose and understanding them is crucial for data manipulation and analysis.

### Merge
Merging datasets involves combining two or more datasets based on a common column or key. This operation is useful when you have related information spread across multiple datasets and you want to consolidate them into a single dataset. The common column serves as the reference point for merging the datasets.

Example:

### Two datasets to merge
df1 = pd.DataFrame({'key': ['A', 'B', 'C'], 'value': [1, 2, 3]})
df2 = pd.DataFrame({'key': ['A', 'B', 'D'], 'value': [4, 5, 6]})

#### Merge based on the 'key' column
merged_df = pd.merge(df1, df2, on='key', how='inner')

------------------------------------------------------------

### Join
Joining datasets involves combining them based on their indices or key columns. Unlike merge, which operates on columns, join operates on indices. Pandas offers the `join()` method for performing this operation.

Example:
#### Two datasets to join:
df1 = pd.DataFrame({'A': [1, 2, 3], 'B': [4, 5, 6]}, index=['a', 'b', 'c'])
df2 = pd.DataFrame({'C': [7, 8, 9], 'D': [10, 11, 12]}, index=['a', 'b', 'd'])

#### Join based on indices:
joined_df = df1.join(df2, how='inner')


------------------------------------------------------------

### Concatenation of columns
Concatenating datasets is useful when you want to combine datasets with the same indices but different columns. This operation appends one dataset's columns to another dataset's columns.

Example:
#### Two datasets with the same indices
df1 = pd.DataFrame({'A': [1, 2, 3], 'B': [4, 5, 6]})
df2 = pd.DataFrame({'C': [7, 8, 9], 'D': [10, 11, 12]})

#### Concatenate columns
concatenated_columns = pd.concat([df1, df2], axis=1)


#### Concatenation of rows
Concatenating datasets along rows is useful when you want to append one dataset's rows to another dataset's rows. This operation stacks datasets on top of each other.

Example:
#### Two datasets with the same indices
df1 = pd.DataFrame({'A': [1, 2, 3], 'B': [4, 5, 6]})
df2 = pd.DataFrame({'C': [7, 8, 9], 'D': [10, 11, 12]})

#### Concatenate rows
concatenated_rows = pd.concat([df1, df2], axis=0)
print(concatenated_rows)

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt  

#Cargar los datasets
wr = pd.read_csv('winequality-red.csv')

wr.head()
```

```python
ww = pd.read_csv('winequality-white.csv')

ww.head()
```

```python
#Imprime el tamaño de del dataset de vinos rojos
print(wr.shape)

#Imprime el tamaño de del dataset de vinos blancos
print(ww.shape)
```

```python
#Concatenación de los datasets
wine = pd.concat([wr, ww], axis = 0)

#Comprobar la dimensón de wine
print("El tamaño del conjunto de datos concatenados es: ", wine.shape)

wine.head()
```

```python
wine.tail()
```

```python
#agrega información estadistica
df_stats = wine.describe()

print(df_stats)
```

```python
#Crear rangos del dataset

rango1 = wine.head(10)
rango2 = wine[300:310]
rango3 = wine.tail(10)
```

```python
wine_rango = pd.concat([rango3, rango1, rango2], axis = 0)
wine_rango
```

```python
#Imprime la media cada uno de los datos
data_mean = wine['free sulfur dioxide'].mean()
print('La media de la columna "free sulfur dioxide" es: ', data_mean)
```

```python
#Imprime la mediana de cada uno de los datos
data_median = wine['free sulfur dioxide'].median()
print('La mediana de la columna "free sulfur dioxide" es: ', data_median)
```

```python
#Imprime la moda de cada uno de los datos
data_mode = wine['free sulfur dioxide'].mode()
print('La moda de la columna "free sulfur dioxide" es: ', data_mode.iloc[0])
```

```python
#Cuartiles

q1 = wine['free sulfur dioxide'].quantile(0.25)
print('El valor del cuartil Q1: ', q1)

q2 = wine['free sulfur dioxide'].quantile(0.50)
print('El valor del cuartil Q2: ', q2)

q3 = wine['free sulfur dioxide'].quantile(0.75)
print('El valor del cuartil Q3: ', q3)
```

```python
#Obtener los valores maximos y minimos
min_value = wine['free sulfur dioxide'].describe().loc['min']
max_value = wine['free sulfur dioxide'].describe().loc['max']

# Calcula el rango
data_range = max_value - min_value
print("\nEl rango de valores es: ", data_range)
```

```python
#standar deviation 
print("La desviación estandar es: ", wine['free sulfur dioxide'].std())
```

```python
#variance
print("La varianza es: ", wine['free sulfur dioxide'].var())
```

```python
# Crear el histograma
plt.hist(wine['free sulfur dioxide'], bins=100, color='pink', edgecolor='black')

# Personalizar el histograma (opcional)
plt.title('Histogram of "free sulfur dioxide"')
plt.xlabel('free sulfur dioxide')
plt.ylabel('Frecuency')

# Mostrar el histograma
plt.show()
```

```python
#Boxplot

# Plotting boxplot of column "free sulfur dioxide"
sns.boxplot(x=wine['free sulfur dioxide'], color = 'green')

# Adding labels and title
plt.xlabel('free sulfur dioxide')
plt.ylabel('Values')
plt.title('Boxplot of column: "free sulfur dioxide" ')

# Display the plot
plt.show()
```

```python
#Obtiene la concatenación de diferentes arhivos en un solo conjunto de datos

data = pd.read_csv("001.csv")
print("Archivos concatenados:\n001.csv")
for i in range (2, 11):
    if i < 10:
        filename = "00"+str(i)
    if i >= 10:
        filename = "0" +str(i)

    file = filename + ".csv"
    print(file)
    dataTemporal = pd.read_csv(file)
    #concatenamos un archivo a la vez
    data = pd.concat([data, dataTemporal], axis = 0)

data
```
