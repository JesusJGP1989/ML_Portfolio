```python
# Importar las librerias
import sklearn.datasets as ds
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.datasets import load_iris
```

```python
iris = load_iris()

# Separamos los datos contenidos en el dataset (diccionario)
# Guardamos los datos del dataset en la variable iris_x
iris_x = iris['data']
# Clases de flores, 0=setosa, 1=versicolor, 2=virginica
iris_y = iris['target']
# Etiquetas de las dimensiones: ancho y largo de los petalos y sepalos
# sepal length (cm)', 'sepal width (cm)', 'petal length (cm)', 'petal width (cm)'
iris_x_labels = iris['feature_names']
# Etiquetas de los nombres de las flores setosa, versicolor, virginica
iris_y_labels = iris['target_names']
```

```python
# Count the occurrences of each type of flower
unique, cuantos = np.unique(iris.target, return_counts=True)

# Grfica circular o de pastel para mostrar las proporcin de los datos
# Describir en una grfica de pie para saber cuantos flores hay de cada tipo
plt.figure(figsize=(6,6))
# Construir la grafica
plt.pie(cuantos, labels=iris_y_labels,
         autopct="%1.1f%%", shadow=True, explode =(0, 0, 0.04))
#Para ver todas las opciones de la librera
plt.show()

# Creamos una segunda representacin de los mismos datos
# Es conveniente presentar la misma informacin con diferentes formatos
plt.figure(figsize=(6,6))
plt.pie(cuantos, labels=iris_y_labels, autopct="%1.1f%%",
        textprops={'fontsize':10},wedgeprops={'edgecolor':'#ffffff'})

#Creamos un circulo para dibujar el donut
circle = plt.Circle(xy=(0, 0), radius=0.7, color='white')
# Agregamos el circulo con el contexto grafico y el mtodo artist
plt.gca().add_artist(circle)
# Agregamos un titulo
plt.title('Class Distribution of Iris Data Points', size=14)
# Mostramos la grfica
plt.show()
```

```python
""""

#### **Diagrama de Violin**
Muestra la distribucin de una variable a traves de la mediana, los cuartiles
 y las pestaas. Pero, adems, nos va a mostrar la densidad o frecuencia de los 
 valores en cada uno de los puntos de su distribucin. Este diagrama nos da una
 idea de los valores clave en cada variable y la distribucin entera para cada
 punto de esa variable.
 
"""""

# Calcula los percentiles

q1, median, q3 = np.percentile(iris_x,[25,50,75] , axis=0)

# Crear un objeto figure
plt.figure(figsize=(8,6))
# Dibuja grfica de violin
parts=plt.violinplot(iris_x, vert=False, showextrema=False)
plt.title('Feature Distribution', size=18)
# Poner etiquetas a cada variable
plt.yticks([1,2,3,4], iris_x_labels)
# Modificamos los violines
for pc in parts['bodies']:
    pc.set_facecolor('#9ec2e6') # cambiar color
    pc.set_edgecolor('#01263a') # cambiar color del contorno
    pc.set_alpha(0.5) # Cambiar la transparencia de los violines
# Usamos la informacin calculada de los cuartiles y mediana
plt.gca().scatter(median, [1,2,3,4], color='white', s=40, zorder=3)
# Creamos la barra de los cuartiles
plt.gca().hlines([1,2,3,4], q1,q3, color='#0a4b78', lw=5)
plt.show()
```

Conlcusion

Bubble plots are effective tools to show information in three dimensions, using x, y and bubble size, it allows data scientist to represent additional information about the data using bubble size to identify trends and outliers. Violin plots are useful to visualize data distribution, comparing multiple groups at the same time, these plots are relevant to compare distributions in different groups of information. 3D plots allow to represent data in three dimensions, they are useful for exploring relationships among three variables identifying patterns and get insights for complex datasets.
