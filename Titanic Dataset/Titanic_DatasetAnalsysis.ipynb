```python
#Imprime el nmero de hombres y mujeres sobrevivientes

print('numero de sobrevivientes:\n{}\n '.format(df['Survived'].value_counts()))
print('numero de mujeres y hombres:\n{}\n'.format(df['Sex'].value_counts() ))
```

```python
#Calcular cuantas mujeres sobrevivieron y que porcentaje representan de la tripulacin

# Sex, survived
ms = df[(df['Sex'] == 'female') & (df['Survived']== 1)] #Filtrar las mujeres sobrevivienes
nms = len(ms) #Obtiene nmero de mujeres sobrevivientes
ntt = len(df) #Obtiene nmero total de pasajeros
pms = (nms / ntt)*100 #Calcula el porcentaje
print('Mujeres sobrevivientes: {}, porcentaje: {:.2f}%\n'.format(nms, pms))
```
